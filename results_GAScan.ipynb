```python
import pandas as pd
import numpy as np
# Leer archivo de texto
with open('result.txt', 'r') as f:
    lines = f.readlines()
```

```python
lines1 = lines[2:247]
lines2 = lines[2143:4189] # x: n1 ... nm c= l= frac
```

```python
# Crear listas vacías para cada columna
col1 = []
col2 = []
col3 = []

# Iterar sobre cada línea del archivo de texto
for line in lines1:
    # Buscar las posiciones de los caracteres ":" y "c="
    pos1 = line.index(":")
    pos2 = line.index("graug=")
    
    # Extraer la cadena entre ":" y "c=" y eliminar espacios en blanco
    col2_value = line[pos1+1:pos2].strip()
    
    # Dividir la línea en las otras cuatro columnas según el patrón de la pregunta
    cols = line.split(' ')
    
    col1.append(cols[0].strip(':'))
    col2.append(col2_value)
    col3.append(cols[len(cols)-1].strip())

# Crear un dataframe con las cinco columnas
df1 = pd.DataFrame({'col1': col1, 'col2': col2, 'col3': col3})
df1['col3'] = df1['col3'].str.replace('graug=', '').astype('float')
df1 = df1.rename(columns={'col3': 'graug'})
print(df1.head())  # Imprime las primeras cinco filas del dataframe
```

```python
pd.value_counts(df1.graug).to_frame().reset_index()
```

```python
# Crear listas vacías para cada columna
col1 = []
col2 = []
col3 = []
col4 = []
col5 = []

# Iterar sobre cada línea del archivo de texto
for line in lines2:
    # Buscar las posiciones de los caracteres ":" y "c="
    pos1 = line.index(":")
    pos2 = line.index("c=")
    
    # Extraer la cadena entre ":" y "c=" y eliminar espacios en blanco
    col2_value = line[pos1+1:pos2].strip()
    
    # Dividir la línea en las otras cuatro columnas según el patrón de la pregunta
    cols = line.split(' ')
    
    col1.append(cols[0].strip(':'))
    col2.append(col2_value)
    col3.append(cols[len(cols)-4])
    col4.append(cols[len(cols)-2])
    col5.append(cols[len(cols)-1].strip())

# Crear un dataframe con las cinco columnas
df = pd.DataFrame({'col1': col1, 'col2': col2, 'col3': col3, 'col4': col4, 'col5': col5})

print(df.head())  # Imprime las primeras cinco filas del dataframe
```

```python
df_bool = df['col3'].str.contains('c=')
df['col4'][df_bool[df_bool==False].index]=df['col5'][df_bool[df_bool==False].index]
df['col5'][df_bool[df_bool==False].index]=None
df['col3'][df_bool[df_bool==False].index]='c=0.000000'
df.loc[df_bool[df_bool==False].index,]
```

```python
df['col3'] = df['col3'].str.replace('c=', '').astype('float')
df['col4'] = df['col4'].str.replace('l=', '').astype('float')
df['col5'] = df['col5'].astype('float')
df = df.rename(columns={'col3': 'c', 'col4': 'l'})
```
